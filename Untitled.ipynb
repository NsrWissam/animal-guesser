```python
import pandas as pd

zoo_names = ['animal','hair','feathers','eggs','milk','airborne','aquatic','predator','toothed','backbone','breathes','venomous','fins','legs','tail','domestic','catsize','type']
df = pd.read_csv('zoo.data', names=zoo_names)
ndf = pd.DataFrame(columns=zoo_names)

for index, row in df.iterrows():
    animal_dict = dict.fromkeys(zoo_names)
    if row['hair']:
        animal_dict['hair'] = ['hair', 'hairy', 'fur','fury']
    if row['feathers']:
        animal_dict['feathers'] = ['feather']
    if row['eggs']:
        animal_dict['eggs'] = ['egg']
    if row['milk']:
        animal_dict['milk'] = ['milk']
    if row['airborne']:
        animal_dict['airborne'] = ['airborne','fly', 'glide','float']
    if row['aquatic']:
        animal_dict['aquatic'] = ['aquatic','water','swim','float']
    if row['predator']:
        animal_dict['predator'] = ['predator','hunt','kill']
    if row['toothed']:
        animal_dict['toothed'] = ['tooth']
    if row['backbone']:
        animal_dict['backbone'] = ['backbone','vertebrate','spine']
    else:
        row['backbone'] = ['invertebrate']
    if row['breathes']:
        animal_dict['breathes'] = ['breath']
    if row['venomous']:
        animal_dict['venomous'] = ['venom','poison']
    if row['fins']:
        if row['fins'] == 1:
            animal_dict['fins'] = [str(row['fins'])+" fin"]
        else:
            animal_dict['fins'] = [str(row['fins'])+" fins"]
    if row['legs']:
        if row['legs'] == 1:
            animal_dict['legs'] = [str(row['legs'])+' leg']
        else: 
            animal_dict['legs'] = [str(row['legs'])+' legs']
    if row['tail']:
        animal_dict['tail'] = ['tail']
    if row['domestic']:
        animal_dict['domestic'] = ['domestic']
    if row['catsize']:
        animal_dict['catsize'] = ['??']
    if row['type']:
        if row['type'] == 1:
            animal_dict['type'] = ['?']
        if row['type'] == 2:
            animal_dict['type'] = ['bird']
        if row['type'] == 3:
            animal_dict['type'] = ['?']
        if row['type'] == 4:
            animal_dict['type'] = ['fish']
        if row['type'] == 5:
            animal_dict['type'] = ['amfibian']
        if row['type'] == 6:
            animal_dict['type'] = ['insect']
        if row['type'] == 7:
            animal_dict['type'] = ['?']
    ndf = ndf.append(animal_dict,True)
ndf
    #a_row = pd.Series([1, 2])
    #df = pd.DataFrame([[3, 4], [5, 6]])
    #row_df = pd.DataFrame([a_row])
    #df = pd.concat([row_df, df], ignore_index=True)
```
